```


# using litecoin
tic= yf.Ticker('LTC-USD')

data = tic.history(period="max", auto_adjust=True)
```

```
df = pd.DataFrame()
df['ds'] = data.index
df['y'] = data['Close'].values
```

```




m = Prophet(daily_seasonality=False)

m.fit(df)
```

```
future = m.make_future_dataframe(730, freq='D')

forecast = m.predict(future)
```

```
m.plot(forecast);
```
